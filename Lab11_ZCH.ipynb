```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
c=pd.read_csv(url)

c
```

```python
cda = c.groupby('iso_code').agg(total_co2 = ('co2', np.sum)).sort_values('total_co2', ascending=False).head(30)
cda
```

```python
cda.plot.bar(figsize=(10, 6))
```
